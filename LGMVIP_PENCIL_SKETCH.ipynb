## Image to Pencil Sketch with Python


## Importing Important Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Importing and Reading Image

```python
img = cv2.imread("Tiger.jpg")
plt.imshow(img)
plt.axis("off") 
plt.show()
```

# Converting to Grayscale Image(Negative Image)

```python
gray=cv2.imread("Tiger.jpg",0)
fig = plt.figure(figsize=(150,70))
plt.imshow(gray,cmap='gray')
```

# Inverting the Gray Image

```python
inverted_image = 255-gray
plt.imshow(inverted_image)
plt.axis("off")
```

# Blurring the Image

```python
#blurring the image by using gaussianBlur function with kernal size (21, 21) 
blurry = cv2.GaussianBlur(inverted_image, (21,21),0)
plt.imshow(blurry)
plt.axis("off")
```

# Creating Pencil Sketch Image

```python
inverted_blurred = 255-blurry
Pencil_Sketch = cv2.divide(gray, blurry, scale = 256.0)
plt.imshow(Pencil_Sketch)
plt.axis("off")
plt.imshow(Pencil_Sketch,cmap='gray')
```
